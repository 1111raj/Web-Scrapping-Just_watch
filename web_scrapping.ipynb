## **Scrapping Movies Data**

```python
def fetch_movie_urls(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
    }
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        return "Failed to retrieve the page, status code:", response.status_code
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup


url = 'https://www.justwatch.com/in/movies?release_year_from=2000'
soup=fetch_movie_urls(url)
print(soup.prettify())

## Hint : Use the following code to extract the film urls
# movie_links = soup.find_all('a', href=True)
# movie_urls = [link['href'] for link in movie_links if '/movie/' in link['href']]

# url_list=[]
# for x in movie_urls:
#   url_list.append('https://www.justwatch.com'+x)
```

## **Scrapping Movie Title**


```python
movie_title_list = []

for url in movie_urls_list:
    soup = fetch_movie_urls(url)
    # Fetch the movie title element
    movie_title_soup = soup.find('h1', class_="title-detail-hero__details__title")
    
    if movie_title_soup:  # Check if the title element exists
        # Extract only the text before the <span> tag
        movie_title = movie_title_soup.contents[0].strip()
    else:
        movie_title = "Title not found"
    
    movie_title_list.append(movie_title)

print(movie_title_list)
```

```python
#scrapping the genre of the movie
movie_genre_list = []

for url in movie_urls_list:
    soup = fetch_movie_urls(url)
    Genre_indi = soup.find('h3', class_='detail-infos__subheading', string='Genres').find_next_sibling('div', class_='detail-infos__value').text.strip()


    if Genre_indi:
        movie_genre = Genre_indi
    else:
        movie_genre = "Genre not found"

    movie_genre_list.append(movie_genre)

for genre in movie_genre_list:
    print(genre)
```

## **Scrapping IMBD Rating**

```python
#scrapping the rating of the movies
movie_rating_list = []

for url in movie_urls_list:
    soup = fetch_movie_urls(url)
    result = soup.find("div", "jw-scoring-listing__rating--group jw-scoring-listing__rating--link")

    if result:
        inner_div = result.find("div")
        if inner_div:
            rating = inner_div.text.split()[0]
            print(rating)
        else:
            print("Inner <div> not found")
    else:
        print("Outer <div> not found")

    movie_rating_list.append(rating)
print(movie_rating_list)
```

## **Scrapping Runtime/Duration**

```python
time_duration_list = []

for url in movie_urls_list:
    soup = fetch_movie_urls(url)
    
    # Find the tag containing "Runtime"
    runtime_heading = soup.find('h3', class_="detail-infos__subheading", string="Runtime")
    if runtime_heading:
        # Find the sibling div with the value
        duration_tag = runtime_heading.find_next_sibling('div', class_="detail-infos__value")
        if duration_tag and duration_tag.text.strip():
            time_duration_list.append(duration_tag.text.strip())
        else:
            time_duration_list.append("Duration not found")
    else:
        time_duration_list.append("Runtime heading not found")

print(time_duration_list)
```

## **Fetching Production Country details**

```python
Country_origin =[]
for url in movie_urls_list:
  response = requests.get(url)
  soup = BeautifulSoup(response.text,"html.parser")
  try:
    country1=soup.find('h3', class_='detail-infos__subheading', string=' Production country ').find_next_sibling('div', class_='detail-infos__value').text.strip()
  except:
    country1 = 'NA'
  Country_origin.append(country1)

print(Country_origin)
```

## **Fetching Production Countries Details**

```python
#scrapping the production company of the movie
production_company_list = []

for url in movie_urls_list:
    soup = fetch_movie_urls(url)
    production_country_head = soup.find('h3', class_='detail-infos__subheading', string=' Production country ')
    if production_country_head:
        production_country = soup.find('h3', class_='detail-infos__subheading', string=' Production country ').find_next_sibling('div', class_="detail-infos__value").text.strip()
        if production_country:
            production_company_list.append(production_country)
        else:
            production_company_list.append("Production company not found")

print(production_company_list)
```

## **Fetching Tv shows Url details**

```python
# Write Your Code here
Tv_shows_url  =[]
Tv_shows_link = soup.find_all("a",class_="title-list-grid__item--link")
for link in Tv_shows_link:
    tv_shows_url = "https://www.justwatch.com"+link["href"]
    Tv_shows_url.append(tv_shows_url)

Tv_shows_url
```

## **Fetching Tv Show Title details**

```python
# Write Your Code here
Tv_shows_Title =[]

for url in Tv_shows_url:
  soup = fetch_movie_urls(url)
  title = soup.find("div",class_="title-detail-hero__content").h1.text.strip().split('(')[0].strip()
  Tv_shows_Title.append(title)

Tv_shows_Title
```

## **Fetching Release Year*

```python
# Write Your Code here
Release_year_ =[]
for url in Tv_shows_url:
  soup = fetch_movie_urls(url)
  year = soup.find("span",class_="release-year").text.strip('()')
  Release_year_.append(year)

Release_year_
```

## **Fetching TV Show Genre Details**

```python
genres = []

for url in Tv_shows_url:
      soup = fetch_movie_urls(url)
      div_tags = soup.find('h3', class_="detail-infos__subheading" ,string= 'Genres').find_next_sibling('div', class_='detail-infos__value').text.strip()
      genres.append(div_tags)

print(genres)
       
```

## **Fetching IMDB Rating Details**

```python
Imdb_rating = []
for url in Tv_shows_url:
    soup = fetch_movie_urls(url)
    span_tags = soup.find_all("span", class_="imdb-score")
    

    for tag in span_tags:
        rating_tag = tag.text
        rating = rating_tag.split()[0]

    if rating:
        Imdb_rating.append(rating)
    else:
        Imdb_rating.append("Rating not found")

print(Imdb_rating) 
```

## **Fetching Streaming Service details*

```python
from bs4 import BeautifulSoup

# Function to extract valid URLs
def extract_links(soup):
    links = []
    for a_tag in soup.find_all("a", class_="title-list-grid__item--link"):
        href = a_tag.get("href")  # Get the 'href' attribute
        if href:
            full_url = f"https://www.justwatch.com{href}"  # Construct full URL if needed
            links.append(full_url)
    return links

# Example soup object
soup = fetch_movie_urls("https://www.justwatch.com/in")
Tv_shows_link = extract_links(soup)  # Now contains a list of valid URLs
```

```python
streaming_providers = []
for url in Tv_shows_url:
    soup = fetch_movie_urls(url)
    streaming_tag = soup.find('div', class_="buybox-row stream")
    
    if streaming_tag:
        stream = streaming_tag.picture.img['alt']
        streaming_providers.append(stream)
    else:
        stream = "Provider not found"
        streaming_providers.append(stream)

print(streaming_providers)
```
